```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')

fig.show()
```

```python
df.info()
```

```python
class_count = df['Class'].value_counts()
class_count
```

```python
class_count.plot(kind='bar')
```

```python
class_0 = df[ df['Class'] == 0]['V1']
class_1 = df[ df['Class'] == 1]['V1']
```

```python
plt.figure()

# Гистограмма для обычных транзакций (Class = 0)
plt.hist(class_0, bins=20, alpha=0.5, color='gray', density=True, label='Class 0')

# Гистограмма для мошеннических транзакций (Class = 1)
plt.hist(class_1, bins=20, alpha=0.5, color='red', density=True, label='Class 1')

plt.xlabel('V1')
plt.ylabel('Плотность распределения')
plt.title('Гистограммы для признака V1')
plt.legend()
plt.show()
```
